```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Uploading the Penn World Table data onto Colab
uploaded = files.upload()
# Reading the Excel file into a Data Frame
df = pd.read_excel("pwt1001.xlsx", sheet_name = 'Data')
```

```python
'''
A) Trend calculation for Nigeria
'''

# Filtering the data for Nigeria and for years 1970 onwards
nigeria = df[(df.iloc[:, 0] == 'NGA')]

# Calculating real GDP per capita through dividing real GDP by population
nigeria['rgdpe_pc'] = nigeria['rgdpe']/nigeria['pop']

def plot_ln_real_gdp(data, title):
    # Plotting the figures
    plt.figure(figsize=(12, 10))

    # Plotting ln real GDP against year
    plt.plot(data['year'], np.log(data['rgdpe']), label='ln Real GDP')
    plt.title(title)  # Adding title
    plt.xlabel('Year')  # Labelling the x axis
    plt.ylabel('ln Real GDP')  # Labelling the y axis
    plt.grid(True)  # Adding a grid for better readability

    # Adding the trend line for ln real GDP
    z = np.polyfit(data['year'], np.log(data['rgdpe']), 1)  # Calculating the coefficients for the linear fit
    p = np.poly1d(z)  # Creating a polynomial function using the coefficients
    plt.plot(data['year'], p(data['year']), color='red', linestyle='--', label='Trend Line')

    plt.legend()  # Adding legend
    plt.tight_layout()  # Adjusting layout for better appearance
    plt.show()

# Calling the function
plot_ln_real_gdp(nigeria, 'ln Real GDP of Nigeria across time')
```

```python
# Plotting ln real GDP per capita against year
def plot_ln_real_gdp_perc(data, title):
  plt.figure(figsize=(12,10))
  plt.plot(data['year'], np.log(data['rgdpe_pc']), label = 'ln real GDP per capita')
  plt.title(title) # Adding title
  plt.xlabel('Year') # Labelling the x axis
  plt.ylabel('ln real GDP per capita') # Labelling the y axis
  plt.grid(True) # Adding a grid for better readability
  # Adding the trend line for ln real GDP per capita
  z_perc = np.polyfit(data['year'], np.log(data['rgdpe_pc']), 1) # Calculating the coefficients for the linear fit
  p_perc = np.poly1d(z_perc) # Creating a polynomial function using the coefficients
  plt.plot(data['year'], p_perc(data['year']), color = 'red', linestyle = '--', label = 'Trend Line')
  plt.legend() # Adding legend
  plt.tight_layout() # Adjusting layout for better appearance
  plt.show() # Displaying the plot

#Calling the function
plot_ln_real_gdp_perc(nigeria, 'ln Real GDP per capita of Nigeria across time')
```

```python
'''
A) Trend calculation for USA
'''

# Filtering the data for USA and for years 1970 onwards
usa = df[(df.iloc[:, 0] == 'USA')]

# Calculating real GDP per capita through dividing real GDP by population
usa['rgdpe_pc'] = usa['rgdpe']/usa['pop']

# Calling the function that plots the graph
plot_ln_real_gdp(usa, 'ln Real GDP of the USA across time')
```

```python
'''
B) Extracting TFP measure for Nigeria
'''

# Here we are going to make use of a Cobb-Douglas production function of the following form:
# rgdpna = A(rnna)^ * (emp)^(1-)
# The benchmark value for  here is  = 0.3
# The values used here are the national accounts-based variables

# Calculating and storing the log of the relevant variables:
nigeria['ln_rgdpna'] = np.log(nigeria['rgdpna'])
nigeria['ln_emp'] = np.log(nigeria['emp'])
nigeria['ln_rnna'] = np.log(nigeria['rnna'])

# Defining the production function
alpha = 0.3
nigeria['alpha_ln_rnna'] = alpha * nigeria['ln_rnna']
nigeria['one_minus_alpha_ln_emp'] = (1-alpha) * nigeria['ln_emp']

# Rearranged the log form of the production function to solve for ln A (ln TFP)
nigeria['ln_tfp'] = nigeria['ln_rgdpna'] - alpha * nigeria['ln_rnna'] - (1-alpha) * nigeria['ln_emp']

# Plotting TFP values over time
plt.figure(figsize=(10, 6))
plt.plot(nigeria['year'], (nigeria['ln_tfp']), label='TFP', color='blue')
plt.title('Nigeria Total Factor Productivity (TFP) Over Time') # Adding title
plt.xlabel('Year') # Labelling the x axis
plt.ylabel('TFP') # Labelling the y axis
plt.legend() # Adding legend
plt.grid(True) # Adding a grid for better readability
plt.show()

# Defining a function that calculates TFP and dlnA for a given value of alpha
def tfp_dlnA_alpha(alpha, df):
  # Creating suitable columns (log transformed)
  df['ln_rgdpna'] = np.log(df['rgdpna'])
  df['ln_rnna'] = np.log(df['rnna'])
  df['ln_emp'] = np.log(df['emp'])
  df['ln_tfp'] = df['ln_rgdpna'] - alpha * df['ln_rnna'] - (1-alpha) * df['ln_emp']
  # Calculating dlnA:
  df['dlnA'] = df['ln_tfp'] - df['ln_tfp'].shift(1)
  tfp_values = df['ln_tfp']
  return df['dlnA']

alpha_values = [0.2, 0.25, 0.3, 0.35, 0.4] # Going to check robustness against these values of alpha
dlnA_values = {} # Creating an empty array that makes the output easier to read later on
# Calculating dlnA for each value of alpha
for alpha in alpha_values: # Iterating through each value of alpha
  df_copy = nigeria.copy() # Creating a copy of the dataframe
  dlnA_values[alpha] = tfp_dlnA_alpha(alpha, df_copy) # Appending the array
  print(f"Alpha = {alpha}, dlnA Values: {dlnA_values[alpha]}:")

```

```python
# Slicing the dataframe to only get the bits relevant to our interval
def filter(df, interval): # Defining the function
  interval_data = df[df['year'] % interval == 0] # Choosing the relevant rows
  last_row = df.iloc[-1]
  interval_data = interval_data.append(last_row, ignore_index=True) # Creating the variable that stores the sub-period data and appending it
  return interval_data

# Producing 5 and 10 year intervals for USA and Nigeria - through calling the function previously defined
nigeria_5 = filter(nigeria, 5)
nigeria_10 = filter(nigeria, 10)
usa_5 = filter(usa, 5)
usa_10 = filter(usa, 10)

#Defining the growth accounting function
def ga(df):
  alpha = 0.3
  # Storing the log change of Y, K, A, and L and putting in the chosen dataframe
  df.loc[:, 'dlogy{}'] = np.log(df['rgdpna']) - np.log(df['rgdpna'].shift(1))
  df.loc[:, 'dloga{}'] = (df['ln_tfp']) - (df['ln_tfp'].shift(1))
  df.loc[:, 'dlogk{}'] = np.log(df['rnna']) - np.log(df['rnna'].shift(1))
  df.loc[:, 'dlogl{}'] = np.log(df['emp']) - np.log(df['emp'].shift(1))
  # The contribution of TFP, K, and L is the change in the log of the respective variable divided by the log change in output
  tfp_contribution = df['dloga{}'] / df['dlogy{}']
  k_contribution = alpha * df['dlogk{}'] / df['dlogy{}']
  l_contribution = (1-alpha) * df['dlogl{}'] / df['dlogy{}']
  return tfp_contribution, k_contribution, l_contribution

# USA - 5 year intervals
tfp_contr_usa_5, k_contr_usa_5, l_contr_usa_5 = ga(usa_5) # Calling the function and storing the values in the relevant variables
print("TFP Contribution (5-year intervals):",tfp_contr_usa_5)
print("Capital Contribution (5-year intervals):",k_contr_usa_5)
print("Labour Contribution (5-year intervals):",l_contr_usa_5)
# Plotting a graph that shows the contribution of TFP, K, and L
plt.figure(figsize=(10, 6))
# Plot TFP Contribution in blue
plt.plot(tfp_contr_usa_5, label='TFP Contribution', color='blue', marker='o')
# Plot Capital Contribution in green
plt.plot(k_contr_usa_5, label='Capital Contribution', color='green', marker='o')
# Plot Labor Contribution in red
plt.plot(l_contr_usa_5, label='Labor Contribution', color='red', marker='o')
plt.legend() # Adding a legend
plt.xlabel('Number of 5 year subperiod (14 subperiods from 1950 to 2020)') # Labelling the x axis
plt.ylabel('Contribution') # Labelling the y axis
plt.title('Contributions to Growth in USA (5-year intervals)') # Adding a title

# USA - 10 year intervals
tfp_contr_usa_10, k_contr_usa_10, l_contr_usa_10 = ga(usa_10)  # Calling the function and storing the values in the relevant variables
print("TFP Contribution (10-year intervals):",tfp_contr_usa_10)
print("Capital Contribution (10-year intervals):",k_contr_usa_10)
print("Labour Contribution (10-year intervals):",l_contr_usa_10)
# Plotting a graph that shows the contribution of TFP, K, and L
plt.figure(figsize=(10, 6))
# Plot TFP Contribution in blue
plt.plot(tfp_contr_usa_10, label='TFP Contribution', color='blue', marker='o')
# Plot Capital Contribution in green
plt.plot(k_contr_usa_10, label='Capital Contribution', color='green', marker='o')
# Plot Labor Contribution in red
plt.plot(l_contr_usa_10, label='Labor Contribution', color='red', marker='o')
plt.legend() # Adding a legend
plt.xlabel('Number of 10 year subperiod (7 subperiods from 1950 to 1970)') # Labelling the x axis
plt.ylabel('Contribution') # Labelling the y axis
plt.title('Contributions to Growth in USA (10-year intervals)') # Adding a title

# Nigeria - 5 year intervals
tfp_contr_n_5, k_contr_n_5, l_contr_n_5 = ga(nigeria_5) # Calling the function and storing the values in the relevant variables
print("TFP Contribution (5-year intervals):",tfp_contr_n_5)
print("Capital Contribution (5-year intervals):",k_contr_n_5)
print("Labour Contribution (5-year intervals):",l_contr_n_5)
# Plotting a graph that shows the contribution of TFP, K, and L
plt.figure(figsize=(10, 6))
# Plot TFP Contribution in blue
plt.plot(tfp_contr_n_5, label='TFP Contribution', color='blue', marker='o')
# Plot Capital Contribution in green
plt.plot(k_contr_n_5, label='Capital Contribution', color='green', marker='o')
# Plot Labor Contribution in red
plt.plot(l_contr_n_5, label='Labor Contribution', color='red', marker='o')
plt.legend() # Adding a legend
plt.xlabel('Number of 5 year subperiod (14 subperiods from 1950 to 2020)') # Labelling the x axis
plt.ylabel('Contribution') # Labelling the y axis
plt.title('Contributions to Growth in Nigeria (5-year intervals)') # Adding a title

# Nigeria - 10 year intervals
tfp_contr_n_10, k_contr_n_10, l_contr_n_10 = ga(nigeria_10) # Calling the function and storing the values in the relevant variables
print("TFP Contribution (10-year intervals):",tfp_contr_n_10)
print("Capital Contribution (10-year intervals):",k_contr_n_10)
print("Labour Contribution (10-year intervals):",l_contr_n_10)
# Plotting a graph that shows the contribution of TFP, K, and L
plt.figure(figsize=(10, 6))
# Plot TFP Contribution in blue
plt.plot(tfp_contr_n_10, label='TFP Contribution', color='blue', marker='o')
# Plot Capital Contribution in green
plt.plot(k_contr_n_10, label='Capital Contribution', color='green', marker='o')
# Plot Labor Contribution in red
plt.plot(l_contr_n_10, label='Labor Contribution', color='red', marker='o')
plt.legend() # Adding a legend
plt.xlabel('Number of 10 year subperiod (7 subperiods from 1950 to 1970)') # Labelling the x axis
plt.ylabel('Contribution') # Labelling the y axis
plt.title('Contributions to Growth in Nigeria (10-year intervals)') # Adding a title



```
